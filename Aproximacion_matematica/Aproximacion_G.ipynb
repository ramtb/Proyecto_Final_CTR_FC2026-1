```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parmetros de Gompertz ---
K = 33500   # Capacidad de Carga
N0 = 1    # Poblacin Inicial
beta = 0.0226 # Tasa de crecimiento y desaceleracin (ajustada para el tiempo de tu curva)
t_max = 1000 # Tiempo mximo

# --- Funcin de Gompertz ---
def crecimiento_gompertz(t, K, N0, beta):
    """Calcula el nmero de clulas N(t) usando el modelo de Gompertz."""
    # Definimos alfa () para simplificar el log(K/N0)
    alfa = np.log(K / N0)

    # La frmula de Gompertz
    N_t = K * np.exp(-alfa * np.exp(-beta * t))
    return N_t

# --- Generacin de Datos ---
pasos_de_tiempo = np.linspace(0, t_max, 100)
cantidad_celulas_gompertz = crecimiento_gompertz(pasos_de_tiempo, K, N0, beta)

# --- Graficacin ---
plt.figure(figsize=(10, 6))

plt.plot(pasos_de_tiempo, cantidad_celulas_gompertz, color='purple', marker='.', linestyle='-', markersize=5, label='Clulas Activas (Modelo de Gompertz)')

# Configuraciones
plt.title('Crecimiento del Tumor (Aproximacin por Modelo de Gompertz)')
plt.xlabel('Paso de Tiempo (t)')
plt.ylabel('Cantidad de Clulas Activas')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='upper left')
plt.ticklabel_format(style='plain', axis='y')

plt.show()
```
