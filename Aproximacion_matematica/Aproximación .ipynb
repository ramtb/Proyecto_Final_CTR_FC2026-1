```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Definición de Parámetros ---
K = 33500   # Capacidad de Carga (límite superior)
N0 = 1    # Población Inicial (t=0)
r = 0.052    # Tasa intrínseca de crecimiento (ajustada para el tiempo de subida)
t_max = 1000 # Tiempo máximo

# --- 2. Función Logística ---
def crecimiento_logistico(t, K, N0, r):
    """Calcula el número de células activas N(t) usando el modelo logístico."""
    C = (K / N0) - 1
    N_t = K / (1 + C * np.exp(-r * t)) #formula logistica
    return N_t

# --- 3. Generación de Datos ---
# Crea un array de 100 puntos de tiempo entre 0 y t_max
pasos_de_tiempo = np.linspace(0, t_max, 100)

# Calcula el número de células para cada paso de tiempo
cantidad_celulas = crecimiento_logistico(pasos_de_tiempo, K, N0, r)

# --- 4. Graficación ---
plt.figure(figsize=(10, 6))

plt.plot(pasos_de_tiempo, cantidad_celulas, color='red', marker='.', linestyle='-', markersize=5, label='Células Activas (Modelo Logístico)')

# Configuraciones de la gráfica.
plt.title('Crecimiento del Tumor (Aproximación matemática)')
plt.xlabel('Paso de Tiempo (t)')
plt.ylabel('Cantidad de Células Activas')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='upper left')

# Formato del eje Y
plt.ticklabel_format(style='plain', axis='y')

# Muestra la gráfica
plt.show()
```
